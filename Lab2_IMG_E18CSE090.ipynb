# Image resizing using bilinear, bicubic and nearest neighbour interpolation methods

```python
linear_img = cv2.resize(image,None,fx=10,fy=10,interpolation=cv2.INTER_LINEAR)
```

```python
plt.subplot(2,2,1)
plt.title("Normal")
plt.imshow(image)

plt.subplot(2,2,2)
plt.title("Bilinear")
plt.imshow(linear_img)
```

```python
cubic_img = cv2.resize(image,None,fx=10,fy=10,interpolation=cv2.INTER_CUBIC)
```

```python
plt.subplot(2,2,1)
plt.title("Normal")
plt.imshow(image)

plt.subplot(2,2,2)
plt.title(" Bicubic")
plt.imshow(cubic_img)
```

```python
nearest_neighbour_img = cv2.resize(image,None,fx=10,fy=10,
                                   interpolation=cv2.INTER_NEAREST)
```

```python
plt.subplot(2,2,1)
plt.title("Normal")
plt.imshow(image)

plt.subplot(2,2,2)
plt.title("Neighbour interpolation methods")
plt.imshow(nearest_neighbour_img)
```
